```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, accuracy_score, confusion_matrix
from sklearn.linear_model import LinearRegression
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pip install plotly==5.5.0
```

```python
data = pd.read_csv("Health_insurance.csv")
```

```python
data1 = pd.get_dummies(data = data, drop_first = True)
```

```python
X = data1.drop(['charges'], axis = 1, inplace = False)
```

```python
Y = data1["charges"]
```

```python
X,Y
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size=0.2, random_state=42)
```

#### Finding base RMSE

```python
base_predictions = np.mean(ytest)
base_predictions
```

```python
base_predictions = np.repeat(base_predictions, len(ytest))
len(base_predictions)
```

```python
base_RMSE = np.sqrt(mean_squared_error(ytest, base_predictions))
print(f"base RMSE {base_RMSE}")
```

```python
type(xtrain),type(xtest),type(ytrain),type(ytest)
```

### Random Forest Regressor

```python
# forest = RandomForestRegressor(oob_score = True)

forest = RandomForestRegressor(n_estimators = 100,
                                                  max_features = 'auto',
                                                  max_depth =100,
                                                  min_samples_split = 10,
                                                  min_samples_leaf = 4,
                                                  random_state = 1,
                                                  oob_score = True)
```

```python
rf_model = forest.fit(xtrain, ytrain)
```

```python
rf_predictions = forest.predict(xtest)
```

```python
rf_predictions
```

```python
rf_mse = mean_squared_error(ytest, rf_predictions)
rf_RMSE = np.sqrt(rf_mse)
print(f"Random Forest RMSE {rf_RMSE}")
```

###### The Root mean squared error is reduced from 12459 to 4578 (approx 60%)

```python
rf_train_r2_value = forest.score(xtrain, ytrain)
rf_test_r2_value = forest.score(xtest, ytest)

print(f"R squared value for training set: {rf_train_r2_value}")
print(f"R squared value for test set: {rf_test_r2_value}")
```



###### training set is able to catch 97.3% variability and test set is able to catch 86.5% variability when RF is instantiated without any details...all default arguments

######  training set is able to catch 90.9% variability and test set is able to catch 87.76% variability when RF is instantiated with details

```python
print(forest.feature_importances_)
print(forest.n_features_in_)
print(forest.base_estimator_)
print(forest.n_outputs_)
print(forest.oob_score_)
print(forest.oob_prediction_)
```

```python
rfresidual = ytest - rf_predictions
```

```python
rfresidual.describe()
```

```python
fig = px.scatter(x = rf_predictions, y= rfresidual, trendline = 'lowess')
fig.show()
```

### Linear Regression Model

```python
linear = LinearRegression(fit_intercept = True)
```

```python
lnrmodel = linear.fit(xtrain, ytrain)
```

```python
lnrpredictions = lnrmodel.predict(xtest)
```

```python
lnr_mse = mean_squared_error(ytest,lnrpredictions)
lnr_RMSE = np.sqrt(lnr_mse)
print(f"Linear Regression RMSE {lnr_RMSE}")
```

```python
lnr_train_r2_value = linear.score(xtrain, ytrain)
lnr_test_r2_value = linear.score(xtest, ytest)

print(f"R squared value for training set: {lnr_train_r2_value}")
print(f"R squared value for test set: {lnr_test_r2_value}")
```

```python

print(linear.coef_)
print(linear.intercept_)
print(linear.singular_)
print(linear.n_features_in_)
print(linear.rank_)
```
